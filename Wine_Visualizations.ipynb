```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

For the purpose of visualization, first red and white wine types will be compared among each other and together.

```python
red_wine = pd.read_csv('red_wine.csv')
```

```python
white_wine = pd.read_csv('white_wine.csv')
```

```python
wines = pd.read_csv('wines.csv')
```

Given the relationship between the components of the wine and the quality. 
Following a analysis through visualization of the ingrediants in both red and white wines and their relationship with their quality label.*italicized text*

VISUALIZATION I
relation by ingrediant and each wine type and/or quality category

```python
#fixed acidity			
ingredient1 = sns.boxplot(x="wine_type", y="fixed acidity", data=wines, hue = 'quality_label')
```

```python
# volatile acidity

ingredient2 = sns.boxplot(x="wine_type", y="volatile acidity", data=wines, hue = 'quality_label')
```

```python
# citric acid

ingredient3 = sns.boxplot(x="wine_type", y="citric acid", data=wines, hue = 'quality_label')
```

```python
# residual sugar

ingredient4 = sns.boxplot(x="wine_type", y="residual sugar", data=wines, hue = 'quality_label')
```

```python
# Residual sugar2
ax = sns.boxplot(x="quality_label", y="residual sugar", data=wines, hue = 'wine_type').set(ylim=(0, 30))
```

```python
# free sulfur dioxide

ax = sns.boxplot(x="quality_label", y="free sulfur dioxide", data=wines, hue = 'wine_type').set(ylim=(0, 100))
```

```python
# total sulfur dioxide

fig, ax = plt.subplots(1,2, figsize=(12, 6))

sns.boxplot(x="wine_type", y="total sulfur dioxide", data=wines, hue = 'quality_label', ax=ax[0]).set_title("total sulfur dioxide")
sns.boxplot(x="quality_label", y="total sulfur dioxide", data=wines, hue = 'wine_type', ax = ax[1]).set_title("total sulfur dioxide")
```

```python
# density
fig, ax = plt.subplots(1,2, figsize=(12, 6))

sns.boxplot(x="wine_type", y="density", data=wines, hue = 'quality_label', ax=ax[0]).set_title("density")
sns.boxplot(x="quality_label", y="density", data=wines, hue = 'wine_type', ax = ax[1]).set_title("density");
sns.catplot(x="wine_type", y="density", hue = 'quality_label', data=wines, kind = 'violin');
sns.catplot(x="quality_label", y="density", hue = 'wine_type', data=wines, kind = 'violin');
```

```python
ax=sns.boxplot(x="quality_label", y="density", data=wines, hue = 'wine_type').set(ylim=(0.98,1.01))
```

```python
# sulphates
fig, ax = plt.subplots(1,2, figsize=(12, 6))

sns.boxplot(x="wine_type", y="sulphates", data=wines, hue = 'quality_label', ax=ax[0]).set_title("sulphates")
sns.boxplot(x="quality_label", y="sulphates", data=wines, hue = 'wine_type', ax = ax[1]).set_title("sulphates");
sns.catplot(x="wine_type", y="sulphates", hue = 'quality_label', data=wines, kind = 'violin');
sns.catplot(x="quality_label", y="sulphates", hue = 'wine_type', data=wines, kind = 'violin');
```

```python
ax = sns.boxplot(x="quality_label", y="sulphates", data=wines, hue = 'wine_type').set(ylim=(0.25,1))
```

```python
# alcohol
fig, ax = plt.subplots(1,2, figsize=(12, 6))

sns.boxplot(x="wine_type", y="alcohol", data=wines, hue = 'quality_label', ax=ax[0]).set_title("alcohol")
sns.boxplot(x="quality_label", y="alcohol", data=wines, hue = 'wine_type', ax = ax[1]).set_title("alcohol");
sns.catplot(x="wine_type", y="alcohol", hue = 'quality_label', data=wines, kind = 'violin');
sns.catplot(x="quality_label", y="alcohol", hue = 'wine_type', data=wines, kind = 'violin');
```
